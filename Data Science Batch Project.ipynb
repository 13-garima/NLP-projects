```python
import gradio as gr              #UI library
from transformers import pipeline        #tranformers pipeline
```

```python
transaltion_pipeline = pipeline("translation_en_to_de")
```

```python
results=transaltion_pipeline("Taj Mahal is in India")
```

```python
results[0]['translation_text']
```

```python
def translate_transformers(from_text):
    results=transaltion_pipeline(from_text)
    return results[0]['translation_text']
```

```python
 translate_transformers('Hugging Face ')
```

```python
interface = gr.Interface(fn=translate_transformers,
                         inputs=gr.inputs.Textbox(lines=2,placeholder="Text to Translate"),
                         outputs ="text")
```
